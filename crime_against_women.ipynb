**Data Preparation and Cleaning**


.So first we are doing our data process i.e. data cleaning.

```python
null_val=crime.isnull().sum()
```

```python
null_val
```

.So there are no null values

Now let's find the total number unique districts where crimes are committed.

```python
districts=crime.DISTRICT.nunique()
```

```python
districts
```

.But in India there are 779 districts.Which means there are messy or false data in large amount

.In this case we better to districts columns and unnamed columns which is not useful for our analysis.

```python
crime.drop(['DISTRICT','Unnamed: 0'],axis=1,inplace=True)
```

```python
crime.rename(columns= {'Kidnapping and Abduction':'Kidnapping_Abduction','Dowry Deaths':
                             'Dowry_Deaths','Assault on women with intent to outrage her modesty':'hurting_of_women_modesty',
                             'Insult to modesty of Women':'Insult_modesty_Women',
                             'Cruelty by Husband or his Relatives':'domestic_cruelty','Importation of Girls':'Importation_Girls'},inplace=True)
```

```python

crime
```

```python
crime['STATE/UT'].unique()
```

.In the 'STATE/UT' column we seen that many names are repeated twice by capital letters and some are repeted with space too,like 'A&N Islands' is repeated  'A & N Islands'.

```python
#remove all repeated upper case values
def remove_upper(r):
  r=r['STATE/UT'].strip()
  r=r.upper()
  return r
crime['STATE/UT']=crime.apply(remove_upper,axis=1)
##Now use replace function to replace the other type of repeated datas as dicussed above
crime['STATE/UT'].replace('A&N ISLANDS','A & N ISLANDS',inplace=True)
crime['STATE/UT'].replace('D&N HAVELI','D & N HAVELI',inplace=True)
crime['STATE/UT'].replace('DELHI UT','DELHI',inplace=True)
```

```python
crime['STATE/UT'].unique()
```

```python
crime['STATE/UT'].nunique()
```

.Let us find the total popualtion where female has victims for the crimes.

```python
crime.columns
```

```python
victim_rapes=crime['Rape'].sum()
victim_kidnappes=crime['Kidnapping_Abduction'].sum()
victim_dowry=crime['Dowry_Deaths'].sum()
victim_hurting=crime['hurting_of_women_modesty'].sum()
victim_insulting=crime['Insult_modesty_Women'].sum()
victim_domestic=crime['domestic_cruelty'].sum()
girls_imported=crime['Importation_Girls'].sum()
```

```python
girls_imported
```

```python
total_populatin_of_girls_victims=victim_rapes+victim_kidnappes+victim_dowry+victim_hurting+victim_insulting+victim_domestic+girls_imported
```

```python
total_populatin_of_girls_victims
```

.The above analysis says that it is an heartbreaking situaton for our womens over 5million females are victims for different crimes like Rapes,importing,domestic_voilence or even death.

.Now lets analyse the all cases visually

```python
fig,axes= plt.subplots(2,3,figsize=(25,12))


axes[0,0].bar(crime['Year'],crime['Rape'],color='b')
axes[0,0].set_xlabel("years")
axes[0,0].set_ylabel("rapes cases per each year")
axes[0,0].set_title("chart of rape cases in India between 2001-2014")


axes[0,1].bar(crime['Year'],crime['Kidnapping_Abduction'],color='black')
axes[0,1].set_xlabel("years")
axes[0,1].set_ylabel("Kidnapping_Abduction cases per each year")
axes[0,1].set_title("chart of Kidnapping_Abduction in India between 2001-2014")

axes[0,2].bar(crime['Year'],crime['Dowry_Deaths'],color='green')
axes[0,2].set_xlabel("years")
axes[0,2].set_ylabel("Dowry_Deaths cases per each year")
axes[0,2].set_title("chart of Dowry_Deaths in India between 2001-2014")

axes[1,0].bar(crime['Year'],crime['hurting_of_women_modesty'],color='cyan')
axes[1,0].set_xlabel("years")
axes[1,0].set_ylabel("hurting_of_women_modesty cases per each year")
axes[1,0].set_title("chart of hurting_of_women_modesty in India between 2001-2014")

axes[1,1].bar(crime['Year'],crime['domestic_cruelty'],color='orange')
axes[1,1].set_xlabel("years")
axes[1,1].set_ylabel("domestic_cruelty cases per each year")
axes[1,1].set_title("chart of domestic_cruelty in India between 2001-2014")

axes[1,2].bar(crime['Year'],crime['Importation_Girls'],color='red')
axes[1,2].set_xlabel("years")
axes[1,2].set_ylabel("Importation_Girls cases per each year")
axes[1,2].set_title("chart of Importation_Girls in India between 2001-2014")


plt.tight_layout()
plt.show()
```

.From the above charts we can analyse the two things,

1.The cases has increased over the years.

2.2014 is the year where the cases against women reported maximum in Rape cases,domestic,voilence etc
